```python
import numpy as np
import pandas as pd

#from src.utils.loggers import logger
#from src.utils.exceptions import CustomException

print("Imported required packages and ready to use........")
```

```python
def load_data(path: str) -> pd.DataFrame:
    """
    Loads the dataset from a CSV file.

    Args:
        path (str): Path to the CSV file

    Returns:
        pd.DataFrame: Loaded dataset
    """
    df = pd.read_csv(path)
    print(f"Dataset loaded successfully with shape {df.shape}")
    return df

data_path = "../data/raw/Telecos_customer_churn.csv"
df = load_data(data_path)
df
```

# there are 4 numerical columns in the dataset:
- the SeniorCitizen column which is an int column
- the tenure column indicating number of months customer has stayed
- the MonthlyCharges column
- the TotalCharges column
but only statistics for three numerical columns are displayed 
observing we can see that the total charges is an object column when its supposed to be numerical so we would change the column to a numerical and recheck the statistic

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.describe()
```

```python
missing = df.isnull().sum()
print(f"Missing values per column:\n{missing}")
```

The TotalCharges Column Contains About 11 Null Values so we should fill them with the multiplication of the monthly charges and the tenure since they correlates

```python
df['TotalCharges'].fillna(df['MonthlyCharges'] * df['tenure'], inplace=True)
df["TotalCharges"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df["Churn"]
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['Churn']
```

# Encoding All The Categorical Columns

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
categorical_cols.remove('customerID')

print(f"Categorical columns to encode: {categorical_cols}")

#One-hot encode categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
print(f"Categorical columns encoded. New shape: {df_encoded.shape}")
```

```python
df_encoded
```

```python
df_encoded.info()
```

```python
#Converting all booleans to int
bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)
df_encoded.info()
```
