```python
import pandas as pd
import numpy as np
```

```python
df_fe = cleaned_df.copy()

df_fe.drop(columns=["customerID"], inplace=True)
```

```python
#Creating A New Column Tenure Groups as a category datatype column
#Customers were grouped into tenure-based lifecycle stages to capture non-linear churn behavior, revealing significantly higher churn risk among early-tenure customers.
df_fe["tenure_group"] = pd.cut(
    df_fe["tenure"],
    bins=[0, 12, 24, 48, 72],
    labels=["0-1yr", "1-2yr", "2-4yr", "4-6yr"]
)
df_fe["tenure_group"]
```

```python
#Then we use one hot encoding to convert to what machine learning models understand as they dont understand category data type
df = pd.get_dummies(df_fe, columns=["tenure_group"], drop_first=True)
df
```

```python
df.info()
```

```python
#Another feature from tenure indicating long term customers only (customers who has spent more that 2yrs)
df["is_long_term"] = (df["tenure"] >= 24).astype(int)
df["is_long_term"]
```

```python
#Creating another feature from monthly charges to show each customers average monthly charge,thats MonthlyCharge \ tenure
df["avg_monthly_charge"] = (
    df["TotalCharges"] / df["tenure"].replace(0, 1)
)
df["avg_monthly_charge"]
```

```python
#Also Creating another column to show high mothly charges flag by determining if the customers monthly charge is greater than the median monthly charge
df["high_monthly_charge"] = (
    df["MonthlyCharges"] > df["MonthlyCharges"].median()
).astype(int)
df["high_monthly_charge"]
```

```python
#Converting the dummy contract columns we earier created to make more sense by adding more weight to 2 years contract and then adding the one year contract 
df["contract_strength"] = (
    df["Contract_Two year"].astype(int) * 2 +
    df["Contract_One year"].astype(int)
)
df["contract_strength"]
```

where 0 marks month - month contract , 1 marks one year contract and 2 marks 2 year contract

```python
#Creating anotehr column to measure and count how embedded a customer is like how much of the services he or she uses
service_cols = [
    "OnlineSecurity_Yes",
    "OnlineBackup_Yes",
    "DeviceProtection_Yes",
    "TechSupport_Yes",
    "StreamingTV_Yes",
    "StreamingMovies_Yes"
]

df["service_count"] = df_fe[service_cols].sum(axis=1)
df["service_count"]
```

WHERE 6 INDICATES A CUSTOMER IS HIGHLY EMBEDDED WITH THE COMPANY AND 1 MEANIS JUST A SHALLOW CUSTOMER 
HAVING IN MIND THAT HIGHKLY EMBEDDED CUSTOMER ARE EXPECTED TO NOT CHURN

```python
df["has_internet"] = (
    df["InternetService_Fiber optic"] | df["InternetService_No"]
).astype(int)
df["has_internet"]
```
